```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition (1).csv')
df
```

```python
cols_to_remove = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours']
df.drop(columns=cols_to_remove, inplace=True)
```

```python
encoder = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = encoder.fit_transform(df[col])
```

```python
features = df.drop('Attrition', axis=1)
target = df['Attrition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.25, random_state=1
)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lr_model = LogisticRegression(max_iter=1200)
lr_model.fit(X_train, y_train)

```

```python
lr_predictions = lr_model.predict(X_test)
```

```python
print("Logistic Regression Accuracy:",
      accuracy_score(y_test, lr_predictions))
print(confusion_matrix(y_test, lr_predictions))
print(classification_report(y_test, lr_predictions))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
nn_model = Sequential()

nn_model.add(Dense(128, activation='relu',
                   input_dim=X_train.shape[1]))
nn_model.add(Dropout(0.25))

nn_model.add(Dense(64, activation='relu'))
nn_model.add(Dropout(0.25))

nn_model.add(Dense(1, activation='sigmoid'))
```

```python
nn_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = nn_model.fit(
    X_train,
    y_train,
    epochs=40,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

```python
nn_predictions = (nn_model.predict(X_test) > 0.5).astype(int)

print("Neural Network Accuracy:",
      accuracy_score(y_test, nn_predictions))
print(confusion_matrix(y_test, nn_predictions))
print(classification_report(y_test, nn_predictions))
```
