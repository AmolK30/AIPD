```python
import base64
import sys

def login():
    password = input("Enter password: ")
    if password != "admin123":
        print("❌ Access Denied")
        return False
    print("✅ Access Granted")
    return True

def encrypt_data(data):
    return base64.b64encode(str(data).encode())

def validate_input(age, income):
    if age < 18 or age > 65:
        return False
    if income < 1000 or income > 100000:
        return False
    return True

def detect_anomaly(age, income):
    if age > 60 and income > 90000:
        return True
    return False

def clean_input(value):
    return abs(value)


if login():

    age = clean_input(int(input("Enter Age: ")))
    income = clean_input(int(input("Enter Monthly Income: ")))

    if not validate_input(age, income):
        print("❌ Invalid input! Possible adversarial attack")
        sys.exit()

    if detect_anomaly(age, income):
        print("⚠️ Anomaly detected! Possible adversarial attack")
        sys.exit()

    enc_age = encrypt_data(age)
    enc_income = encrypt_data(income)

    print("\n🔐 Encrypted Data:")
    print("Age:", enc_age)
    print("Income:", enc_income)

    print("\n✅ Data is protected from breaches")
```
